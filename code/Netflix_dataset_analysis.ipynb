```python
import matplotlib.pyplot as plt
```

```python
# 2000년 이후 데이터만 추출
netflix_data = data[data['release_year'] >= 2000]

# 연도별 제작한 작품 수 구하기
tv_shows = netflix_data[netflix_data['type'] == 'TV Show']
movie = netflix_data[netflix_data['type'] == 'Movie']

tv_shows_count = tv_shows.groupby('release_year')['title'].count()
movie_count = movie.groupby('release_year')['title'].count()

# 시각화
plt.plot(tv_shows_count.index, tv_shows_count.values, marker='o', label='TV Shows')
plt.plot(movie_count.index, movie_count.values, marker='o', label='Movies')
plt.xlabel('Year')
plt.ylabel('Number of releases')
plt.legend()
plt.show()
```

```python
# date_added 열을 datetime 형식으로 변경
netflix_data['date_added'] = pd.to_datetime(data['date_added'])

# date_added 열에서 연도 정보만 추출
netflix_data['year_added'] = netflix_data['date_added'].dt.year

# 영화와 TV Show 구분
movies = netflix_data[netflix_data['type'] == 'Movie']
tv_shows = netflix_data[netflix_data['type'] == 'TV Show']

# 각 연도별 영화와 TV Show 개수 구하기
movies_count = movies['year_added'].value_counts().sort_index()
tv_shows_count = tv_shows['year_added'].value_counts().sort_index()

# 시각화
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(movies_count.index, movies_count.values, marker='o', label='Movies')
ax.plot(tv_shows_count.index, tv_shows_count.values, marker='o', label='TV Shows')
ax.set_xlabel('Year')
ax.set_ylabel('Number of additions')
ax.set_title('Netflix additions by year')
ax.legend()
plt.show()
```

* 2016년도부터 공개량이 폭발적으로 증가
* 공개되는 영화는 2019년도부터 감소세

```python
# listed_in 열에서 ,로 구분되는 항목 추출
listed_in_list = data['listed_in'].str.split(',').tolist()

# 중복 제거
unique_listed_in = set([item.strip() for sublist in listed_in_list for item in sublist])

# 개별 출력
for item in unique_listed_in:
    print(item)
print(len(unique_listed_in))
```

```python
# listed_in 열에서 ','로 구분된 항목들을 분리하여 리스트로 저장
listed_in = data['listed_in'].str.split(',').apply(pd.Series)

# 분리된 항목들을 개별적으로 추출하여 중복되는 항목들의 개수 계산
listed_in_counts = listed_in.apply(pd.Series.value_counts).fillna(0)

# 각 항목들을 더하여 총 개수 계산
listed_in_counts['total_count'] = listed_in_counts.sum(axis=1)

# 총 개수를 기준으로 정렬
listed_in_counts_sorted = listed_in_counts.sort_values(by='total_count', ascending=False)

# 출력
print(listed_in_counts_sorted["total_count"].head(20))
```

* International Movies가 가장 많이 등장한 장르로 2624건 있었으며, 그 다음으로 Dramas, Comedies, Action & Adventure, Documentaries 등이 많이 등장하는 장르다.
* 장르는 작품에 따라 복수 선택 가능

```python
# 결측치 처리
netflix_data['duration'] = data['duration'].fillna('0')  # 결측치를 '0'으로 채우기

# 분 단위의 상영 시간을 정수형으로 변환
netflix_data['duration'] = netflix_data['duration'].apply(lambda x: int(x.split(' ')[0]))

# 히스토그램 그리기
plt.hist(netflix_data['duration'], bins=50)
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Titles')
plt.show()
```

* 90분에서 120분 사이에 집중적으로 분포하는 것으로 나타남
* 0분은 결측치
* 200분을 초과하는 작품도 드물게 있음
